# Helper Functions

```python
# Helper Functions
def p(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'p%s' % n
    return percentile_

pd.options.display.float_format = '{:40,.3f}'.format
```
